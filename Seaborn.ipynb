**Aula 5 - PairGrids**

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
#espÃ©cies de flores:
```

```python
iris['species'].value_counts()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
#grÃ¡ficos de dispersÃ£o utilizando todas as variÃ¡veis numÃ©ricas possÃ­veis, semelhante ao pairplot
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris, hue='species')
#pelo pairplot terÃ­amos:
```

**Aula 6 - Estilos e cores**

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
#padrÃ£o
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
#detalhes a esquerda
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('dark')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine()
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(left=True)
```

```python
sns.set_style('ticks')
plt.figure(figsize=(12,8))
sns.countplot(x='sex', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, height=2, aspect=3)
```

```python
sns.set_context('poster')
sns.lmplot(x='total_bill', y='tip', data=tips)
#parÃ¢metros permitidos: paper, poster, notebook, talk
```

```python
sns.set_context('poster', font_scale=0.4)
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='inferno')
```
